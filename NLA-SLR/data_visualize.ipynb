```python
import pickle

data_train_path = '../../data/wlasl_2000/train_new.pkl'
data_dev_path = '../../data/wlasl_2000/dev_new.pkl'
data_test_path = '../../data/wlasl_2000/test_new.pkl'

with open(data_train_path, 'rb') as f:
    data_train_new = pickle.load(f)
with open(data_test_path, 'rb') as f:
    data_test_new = pickle.load(f)
with open(data_dev_path, 'rb') as f:
    data_dev_new = pickle.load(f)    
```

```python
import cv2
```

```python
frame_width_list = []
frame_height_list = []

for file in data_train_new:
    path ='/data/wlasl_2000/'+file['video_file']
    #print(path)
    #cap = cv2.VideoCapture(os.path.join(path, file))
    cap = cv2.VideoCapture(path)
    frame_width, frame_height = int(cap.get(3)), int(cap.get(4))
    frame_width_list.append(frame_width)
    frame_height_list.append(frame_height)

for file in data_test_new:
    path ='/data/wlasl_2000/'+file['video_file']
    #print(path)
    #cap = cv2.VideoCapture(os.path.join(path, file))
    cap = cv2.VideoCapture(path)
    frame_width, frame_height = int(cap.get(3)), int(cap.get(4))
    frame_width_list.append(frame_width)
    frame_height_list.append(frame_height)
for file in data_dev_new:
    path ='/data/wlasl_2000/'+file['video_file']
    #print(path)
    #cap = cv2.VideoCapture(os.path.join(path, file))
    cap = cv2.VideoCapture(path)
    frame_width, frame_height = int(cap.get(3)), int(cap.get(4))
    frame_width_list.append(frame_width)
    frame_height_list.append(frame_height)
```

```python
from matplotlib import pyplot as plt
plt.hist(frame_width_list, 100)

plt.show()
```

```python
plt.hist(frame_height_list, 100)

plt.show()


```

```python
for file in data_train_new:
    path ='/data/wlasl_2000/'+file['video_file']   
    cap = cv2.VideoCapture(path)
    resize =256
    dest_path = '/data/wlasl_2000/wlasl_resized/'+file['video_file']

    frame_width, frame_height = int(cap.get(3)), int(cap.get(4))


    out = cv2.VideoWriter(dest_path, cv2.VideoWriter_fourcc(
        'm', 'p', '4', 'v'), 30, (resize, resize))



    while cap.isOpened():
        success, image = cap.read()
        if not success:
            print("Ignoring empty camera frame.")
            # If loading a video, use 'break' instead of 'continue'.
            break
        #ret, frame = cap.read()
        img = cv2.resize(image, (resize, resize))

        out.write(img)
    cap.release()
    out.release()


```
